```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
from stemgraphic import stem_graphic
```

```python
df = pd.read_csv('bear_info.csv')
df.describe()
```

```python
stem_graphic(df['Age'])
```

```python
data = df['Month']
bins = np.arange(4,13)-0.5

hist, edges = np.histogram(data, bins=bins)

y = np.arange(1,hist.max()+1)
x = np.arange(4,12)
X,Y = np.meshgrid(x,y)

plt.scatter(X,Y, c=Y<=hist, cmap="Greys")
plt.xlabel('Month of Measurement',fontsize=15)
plt.ylabel('No. of Bears',fontsize=15)
plt.title('Bear Month of Measurement',fontsize=15)

plt.show()
```

```python
data = df['Sex']
bins = np.arange(5)-0.5

hist, edges = np.histogram(data, bins=bins)

y = np.arange(1,hist.max()+1)
x = np.arange(4)
X,Y = np.meshgrid(x,y)

plt.scatter(X,Y, c=Y<=hist, cmap="Greys")
plt.xticks(range(1,3), ['Male', 'Female'])

plt.xlabel('Sex Code',fontsize=15)
plt.ylabel('No. of Bears',fontsize=15)
plt.title('Bear Sex',fontsize=15)
plt.show()
```

```python
data = df['Headlength']
plt.hist(data, bins=np.arange(min(data),max(data)+2)-0.5, ec="k")

plt.xticks(range(9,18))
plt.xlabel('Headlength (in.)',fontsize=15)
plt.ylabel('No. of Bears',fontsize=15)
plt.title('Bear Headlength',fontsize=15)
plt.show()
```
